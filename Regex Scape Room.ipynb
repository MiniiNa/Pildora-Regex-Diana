# 🔐 Nivel 1: Escáner de códigos

🛠️ Herramienta desbloqueada: `findall()`

Esta función sirve para:
👉 encontrar MUCHAS cosas a la vez

El hacker ha dejado muchos códigos secretos escondidos.

Tu misión:
👉 Encuentra TODOS los códigos secretos

```python
import re

texto = "\n".join(logs)

patron = r""

resultados = re.findall(patron, texto)

print("🎉 ¡Genial! Has encontrado los códigos secretos:", resultados)
print("🔓 Nueva pista desbloqueada")
```

# 🔐 Nivel 2: La IP sospechosa

🛠️ Herramienta desbloqueada: `search()`

Esta función sirve para:
👉 encontrar SOLO la primera coincidencia

El sistema detectó una IP extremadamente sospechosa…

💡 Pista:
Una IP imposible (como 999.999.999.999)

Tu misión:
👉 Encuentra ESA IP

```python
patron = r""

resultado = re.search(patron, "\n".join(logs))

print("🚨 ¡Alerta! Has identificado una IP sospechosa:", resultado.group() if resultado else "No encontrada")
print("👀 Esa IP no debería existir...")
```

# 🔐 Nivel 3: Separando pistas

🛠️ Herramienta desbloqueada: `split()`

Esta función sirve para:
👉 dividir texto en partes

Vamos a analizar una línea de log:

👉 "user:ana ip:192.168.1.10 action:login_failed"

Tu misión:
👉 Separar la línea en partes usando regex

```python
linea = logs[0]

patron = r""

partes = re.split(patron, linea)

print("✂️ ¡Perfecto! Has dividido la información correctamente.", partes)
```

# 🔐 Nivel 4: Limpiando datos

🛠️ Herramienta desbloqueada: `sub()`

Esta función sirve para:
👉 reemplazar cosas en el texto

El sistema está lleno de ruido 😵

Tu misión:
👉 Elimina la palabra "ERROR" de los logs

```python
texto = "\n".join(logs)

patron = r""
nuevo_texto = re.sub(patron, "", texto)

print(nuevo_texto[:500], "🧹 ¡Datos limpios!")  # solo mostramos parte
```
